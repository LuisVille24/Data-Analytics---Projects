```python
# Height Adjustments here
# Converting to CM

def height_conv(h):
    
    feet_index = h.index("'")
    inches_index = h.index("\"")
    
    feet = int(h[:feet_index])
    inches = int(h[feet_index + 1:inches_index])
    
    h_cm = (feet * 30.48) + (inches * 2.54)
    
    return np.round(h_cm)
    
# df1['Height'] = df1['Height'].apply(height_conv) --- DONE
df1.rename(columns={'Height': 'Height(cm)'}, inplace=True)
```

```python
# Loan Date End investigation 
# Column is null for the players that are not on loan. It shows the end of loaned players' contracts.
# Having NULL values for this column is unavoidable at this point of the ETL process. 
# Keep a close watch on this column during EDA. 

df1[df1['Loan Date End'].isnull() == False]['Loan Date End']
```

```python
# Value = Market_Value(€)

# Multiply the subcripts and extract the € sign for the column title


def conversion(value):
    final_value = ''
    for char in value:
        if char.isdigit() or char == '.':
            final_value = final_value + char # string
    
    for i in value:
        if i == 'M':
            return float(final_value) * 1000000
        elif i == 'K':
            return float(final_value) * 1000
        
    return float(final_value)

        
# df1['Value'] = df1['Value'].apply(conversion) -- DONE
# df1.rename(columns={'Value': 'Market_Value(Euro)'}, inplace=True) -- DONE


# As a result of the function above, players with no Market Value were assigned NULL instead of 0 dollars. Let's change that:

df1['Market_Value(Euro)'] = df1['Market_Value(Euro)'].fillna(0)
df1['Market_Value(Euro)'].head()
```
